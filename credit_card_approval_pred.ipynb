# Credit Card Approval Prediction

In the context of credit card approval prediction, financial institutions receive applications from individuals seeking credit

cards. Each application is associated with a set of features or attributes that provide information about the applicant's

financial history, income, employment status, debt, and other relevant factors. The goal is to analyze these features and

predict whether the applicant is a good or bad credit risk.

### Objective


- The primary objective is to build a predictive model that can assist financial institutions in automating the credit approval
  process.
  
  The model aims to:

- Assess Creditworthiness: Evaluate the creditworthiness of an individual based on historical and current financial information.

This involves analyzing factors such as income, outstanding debt, credit history, and other relevant features.

- Minimize Risk: Minimize the risk of approving credit to individuals who are likely to default on payments. The model should

identify potential high-risk applicants to prevent financial losses for the institution.

- Improve Decision-Making: Provide a data-driven approach to credit approval decisions, enhancing the efficiency and consistency

of the decision-making process.

- Compliance: Ensure compliance with regulatory requirements and fair lending practices by making objective, non-discriminatory

decisions based on relevant factors.

- Customer Experience: Enhance the overall customer experience by providing quicker and more consistent credit decisions.

### Machine Learning Approach:



- To achieve the objective, a machine learning model can be trained using historical credit data. Common machine learning

algorithms for this task include logistic regression, decision trees, random forests, and support vector machines. The model is

trained on a labeled dataset where the labels indicate whether each credit application was approved or denied.


- After training, the model can be tested on a separate dataset to evaluate its performance and fine-tuned for optimal results.

The ultimate goal is to deploy the model into a credit approval system, where it can analyze new credit card applications and

provide predictions based on the applicant's information.


- Building a credit card approval prediction model involves a combination of data preprocessing, feature engineering, model

training, and evaluation. Additionally, ethical considerations must be taken into account to ensure fairness and prevent bias in

the credit approval process.

### Hypothesis

When developing a predictive model for credit card approval using machine learning, we can formulate various hypotheses to guide analysis and feature selection. Hypotheses help us make informed decisions about which features to include, how to preprocess the data, and which algorithms to use. Here are some hypotheses we can consider:


- Income Hypothesis:

Hypothesis: Applicants with higher incomes are more likely to have their credit card applications approved.
Rationale: Higher incomes generally indicate a greater ability to repay credit card debts.

- Employment Status Hypothesis:

Hypothesis: Employed applicants are more likely to have their credit card applications approved.
Rationale: Employed individuals often have a more stable source of income, making them potentially less risky.

- Credit History Hypothesis:

Hypothesis: Applicants with a positive credit history are more likely to have their credit card applications approved.
Rationale: Past credit behavior is a strong predictor of future behavior, and a positive credit history is indicative of responsible financial management.

- Number of Dependents Hypothesis:

Hypothesis: Applicants with fewer dependents are more likely to have their credit card applications approved.
Rationale: Having fewer dependents may mean lower financial responsibilities and a higher capacity to manage additional credit.

- Age Hypothesis:

Hypothesis: Younger applicants may be less likely to have their credit card applications approved.
Rationale: Younger individuals might have limited credit history and less stable financial situations.

- Feature Interactions Hypothesis:

Hypothesis: Interactions between certain features (e.g., credit score and income) may have a significant impact on the prediction.
Rationale: The combined effect of certain features may be more influential than the individual effects.

It's essential to test these hypotheses during the data exploration and analysis phase and refine them based on the insights gained from the dataset. Additionally, be mindful of potential biases and ethical considerations to ensure fair and unbiased credit decisions.

```python
Q1 = np.percentile(clean_df['Annual_income'], 25)
Q2 = np.percentile(clean_df['Annual_income'], 50)
Q3 = np.percentile(clean_df['Annual_income'], 75)
```

```python
# calculating inter quartile range
IQR = Q3 - Q1
print('Interquartile range is', IQR)

# lower limit
low_lim = Q1 - 1.5 * IQR
# upper limit
up_lim = Q3 + 1.5 * IQR

print('low_limit is', low_lim)
print('up_limit is', up_lim)

outlier =[]
for x in clean_df['Annual_income']:
    if ((x> up_lim) or (x<low_lim)):
         outlier.append(x)
print(' outlier in the dataset is', outlier)
```

```python
clean_df['Annual_income']=clean_df['Annual_income'].clip(lower=low_lim,upper=up_lim)
```
